```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

df=pd.read_csv("Expanded_data_with_more_features.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df=df.drop("Unnamed:0",axis=1)
print(df.head())
#removing unwanted coloumn
```

# Transforming weekly study hours coloumn

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

# Gender Distribution

```python
plt.figure(figsize=(3,4))
plt.title("Gender Distribution")
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

### We can see from the above count plot that females in the data are slightly more than males.

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

### We can see from the above heat map that parent's education plays a vital role in children marks.

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot=True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

### From the above chart we can conclude that Parent's Marital Status doesn't play any significant role in their child's marks

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()
l=["group A","group B","group C","group D","group E",]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Ethnic Group")
plt.show()
```

### GROUP C have the highest proprtion in distribution

```python
gb2=df.groupby("WklyStudyHours").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb2)
plt.figure(figsize=(5,5))
sns.heatmap(gb2,annot=True)
plt.title("Relationship between WeeklyStudyHours and Student's Score")
plt.show()
```

### We can conclude from above chart that higher the weekly study hours higher the average marks scored by student
